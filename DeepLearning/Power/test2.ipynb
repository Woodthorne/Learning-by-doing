```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# Skapa en tidsindexerad dataserie
np.random.seed(0)
dates = pd.date_range(start='2020-01-01', periods=100, freq='M')
data = np.random.normal(loc=0.5, scale=2, size=100).cumsum() + 10 * np.sin(np.linspace(0, 20, 100))
df = pd.DataFrame(data, index=dates, columns=['value'])

# Utfr en ssongsmssig dekomposition
result = seasonal_decompose(df['value'], model='additive')
# result = seasonal_decompose(df['value'], model='multiplicative')

# Plotta dekompositionsresultatet
result.plot()
plt.show()
```
