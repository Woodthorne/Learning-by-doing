```python
import pandas as pd
import glob

def las_in_csv_mapp(mapp_vag):
    # Skapar en tom lista för att lagra DataFrames
    dataframes_lista = []
    
    # Använder glob för att hitta alla csv-filer i mappen
    for fil_vag in glob.glob(mapp_vag + '/*.csv'):
        # Läser in csv-filen som en DataFrame och lägger till den i listan
        df = pd.read_csv(fil_vag)
        dataframes_lista.append(df)
    
    return dataframes_lista

# Exempelanvändning
mapp_vag = 'din_mapp_vag_här'
df_lista = las_in_csv_mapp(r'C:\Users\Sebastian\Systementor\TH_PYAI23\Maskininlarning_teori_och_praktisk_tillampning\Lektioner\PVSPYH23Lektion10\Fun')

# Du kan nu arbeta med df_lista som innehåller alla dina DataFrames
```

```python
def plot_fun_data(df_list):
    for index, df in enumerate(df_list):
        plt.figure(figsize=(8,6))
        plt.scatter(df['x'], df['y'], c=df['color'])
        plt.title(f'Scatter plot för DataFrame {index+1}')
        plt.xlabel('x-värde')
        plt.ylabel('y-värde')
        plt.show()


plot_fun_data(df_lista)
```
